## To prove a number to be probably prime

```python
from random import randint

def Fermat_isProbablyPrime(n,k = 5):
    if (n < 2):
        return False
    output = True
    for i in range(0, min(n,k)):
        a = randint(1, n-1)
        if (pow(a, n-1, n) != 1):
            return False
    return output
```

```python
# max. Anzahl Zahlen, welche geprüft werden sollen, ob sie Primzahlen sind
max = 2000

for n in range(2,max):
    if Fermat_isProbablyPrime(n,10):
        print("n = %d is prime" % n) 
```

# Square and Multiply Algorithmus

```python
# Square and Multiply
# Attributes: a^e modulo m
def squareAndMultiply(a,e,m):
    # Convert e to its binary form
    binexp = []
    while e!= 0:
        binexp.append(e % 2)
        e = e // 2
    res = 1
    binexp.reverse()
    for i in binexp:
        if i == 0:
            res = (res*res) % m
        else:
            res = (res*res*a) % m
        print(i,"\t",res)
    return res
```

# Miller-Rabin Primality Test

```python
from random import randrange

def MillerRabin(p):
    d = p-1
    r = 0
    while d % 2 == 0:
        d //= 2
        r += 1
    a = randrange(2,p-1)
    x = (a ** d) % p
    if x == 1 or x == p-1:
        return True
    while r > 1:
        x = (x * x) % p
        if x == 1:
            return False
        if x == p-1:
            return True
        r -= 1
    return False
```
